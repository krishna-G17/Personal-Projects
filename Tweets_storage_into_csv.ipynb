## getting stream of tweets

```python
consumer_key="rP59j8owdwI2u9T4UyDXkjm9H"
consumer_secret = "aAiEhwZ94qaVL5Wsty5iL4oEAyrPGC5kFt8HOWSxfm2MumP8Ic"
access_token = "1107457263290056704-SRY8G5RIcQk62g9bCjRPFmd3oeddy1"
access_token_secret = "SYzDNBw0hBwEkrVpYnqVb7l3vgSHB1EeJIucG5gHDoIek"
```

```python

import tweepy
from tweepy import Stream
from tweepy import StreamListener
import json
from textblob import TextBlob
```

```python
import re
import csv
```

```python
# creating a listener

trump = 0
joe = 0

header_name = ['Trump' ,'Joe Biden']
with open('sentiment.csv','w') as file:
    writer =csv.DictWriter(file,fieldnames = header_name)
    writer.writeheader()


class listener(StreamListener):
    
    def on_data(self,data):
        raw_tweets = json.loads(data)
        try:
            tweets = raw_tweets['text']
            tweets = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9a-zA-Z \t])|(\w+:\/\/\s+)","",tweets).split())
            # reomving retweets 
            tweets = ' '.join(re.sub('RT','',tweets).split())
            blob = TextBlob(tweets.strip())
            
            global trump
            global joe
            trump_sentiment = 0
            joe_sentiment = 0
            
            for sent in blob.sentences:
                if "Trump" in sent and "Joe Biden" not in sent:
                    trump_sentiment = trump_sentiment + sent.sentiment.polarity
                else:
                    joe_sentiment = joe_sentiment + sent.sentiment.polarity
                    
            trump = trump + trump_sentiment
            joe = joe + joe_sentiment
            
            with open('sentiment.csv','a') as file:
                writer = csv.DictWriter(file, fieldnames = header_name)
                info = {
                    'Trump':trump,
                    'Joe Biden': joe
                }
                writer.writerow(info)
                    
            print(tweets)
            print()
        except:
            print('error')
        
        
    def on_error(self,status):
        print(status)
```

```python
auth = tweepy.OAuthHandler(consumer_key,consumer_secret)
auth.set_access_token(access_token,access_token_secret)
```

```python
twitter_stream = Stream(auth, listener())
twitter_stream.filter(track=['Trump','Joe Biden'])
```
